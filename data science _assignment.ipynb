# To find similarity between two columns
# if find similarty between two columns using T-test .it used analysis of only two columns
# it has two type independent (unpair) and dependent (pair)
# unpair -different group with respective one column
# pair - same group between different two column analysis

```python
sample.head()
```

```python
from scipy.stats import  ttest_ind
degree=sample[sample['degree_t']=='Sci&Tech']['salary']
sp=sample[sample['specialisation']=='Mkt&HR']['salary']
ttest_ind(degree,sp)
```

# covert normal distribution into standard distribution
# different type unit convert into one type so easy to analysis 

```python
#convert normal distribution to standard distribution for salary
def stdnormalDist(sample):
    mean=sample.mean()
    std=sample.std()
    value=[i for i in sample]
    z_score=[((j - mean)/std) for j in value]
    sns.distplot(z_score,kde=True)
```

```python
stdnormalDist(sample['salary'])
```

```python
sns.distplot(sample['salary'])
```

# probablites density function find different range between 600000 to 700000


```python
def get_pro_density_fn(sample,start,end):
    from scipy.stats import norm
    import matplotlib.pyplot as plt
    ax=sns.distplot(sample,kde=True ,kde_kws={'color':'red'},color='green')
    plt.axvline(start,color='red')
    plt.axvline(end,color='red')
    sample1=sample
    sample_mean=sample1.mean()
    sample_std=sample1.std()
    print("mean&std",(sample_mean,sample_std))
    dist=norm(sample_mean,sample_std)
    values=[value for value in range (start,end)]
    probablities=[dist.pdf(value) for value in values]
    prob=sum( probablities)
    print("The area betwwen range ({}{}:{}".format(start,end,sum( probablities)))
    return prob
          
    
```

```python
get_pro_density_fn(sample['salary'],600000,700000)
```
